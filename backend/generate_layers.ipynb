## Get First 50 Stocks from NSE

```python
import io
import pandas as pd
import requests
from enum import Enum

# 1. Fetch Top 50 (Nifty 50) companies directly from NSE
url = 'https://archives.nseindia.com/content/indices/ind_nifty50list.csv'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
df = pd.read_csv(io.StringIO(response.text))

nse_symbols = df['Symbol'].tolist()

import yfinance as yf

def is_valid_ticker(symbol):
    ticker = yf.Ticker(symbol)
    
    # Try fetching recently available data
    history = ticker.history(period="1d")
    
    # If the dataframe is not empty, the symbol is valid
    return not history.empty

yf_symbols = []
num_symbols = 100
for symbol in nse_symbols[:num_symbols]:
    yf_symbol = f"{symbol}"
    yf_symbols.append(yf_symbol)

# Example usage:
print(f"First {num_symbols} YFinance Symbols: {yf_symbols[:num_symbols]}")
```
